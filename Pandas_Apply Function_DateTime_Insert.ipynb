```python
data = {'Cust_ID': ['C1051','C1005','C1050','C2500'],
        'Principal': [200000, 400000, 250000, 300000],
        'RoI': [8.90, 10.45, 12.60, 9.50],
        'Tenure': [3,7,4,5]}
df = pd.DataFrame(data, index=list('1234'))
print('Given Dataset:')
print(df)
"""Simple Interest = P*T*R/100
Compount Interest = P(1 + R/100)^N"""

df['SI'] = (df['Principal'] * df['Tenure'] * df['RoI'])/100
df['CI'] = (df['Principal']*(1+df['RoI']/100)**df['Tenure']-df['Principal']).round()

print('DataFrame with SI and CI:')
print(df)
```

```python
df.sort_values('SI', ascending=False)                         # Sorted by 'SI' in descending order
```

```python
df.sort_values('CI')                         # sorted by 'CI'
```

```python
# Check the Customer paying Max Simple Interest and Max Compound Interest
df.groupby('SI').get_group(df['SI'].max())  # Record of customer paying Max Simple Interest
```

```python
df.groupby('CI').get_group(df['CI'].min())  # Record of customer paying min Compound Interest
```

```python
# both min SI, min CI paying customers information in one DataFrame

result = [df['SI'].min(), df['CI'].max()]
df.loc[df['SI'].isin(result) |df['CI'].isin(result)]
```

#### Converting  object Date to DateTime type

```python
data = {"Cust ID": ['C1001','C1005','C1050','C2500'],
        "Principal": [20000,40000,25000,30000],
       'Rate' : [8.9,10.45,7.6,9.5],
       'Time' : [3,4,5,2],
       'Date' : ['17-05-2023','17/03/2022','10-11-2019','06/08/2020']}

df2 = pd.DataFrame(data)
df2
```

```python
df2.dtypes          # Date Column is object type not datetime type
```

```python
from datetime import datetime
def clean_date(x):
    try:
        new_date = datetime.strptime(x, '%d-%m-%Y')
    except:
        new_date = datetime.strptime(x, '%d/%m/%Y')
    return new_date

df2['Date'] = df2['Date'].apply(clean_date)
df2
```

```python
df2.dtypes                   # Date column converted into datatime object
```

### Insert Function

```python
To insert particular column at specified location of DataFrame


Syntax:
    df.insert(loc = int, column = 'Col_Name', value, allow_duplicates=True/False)


Parameters:
    loc : int
        Insertion index. Must verify 0 <= loc <= len(columns)
        
    column : str, number, or hashable object
        Label of the inserted column.
            hachable object    : Immutable (numeric, str, tuple, frozenset, bool)
            un-hashable object : Mutable (list, set)
            
    value : Scalar, Series, or array-like
        
    allow_duplicates : bool, optional default False
        False >> will not insert duplicate column
        True  >> will insert duplicate column


Note: Duplication of columns is useless
    Raises a ValueError if `column` is already contained in the DataFrame,
    unless `allow_duplicates` is set to True.
```

#### Inserting two columns at a time at specific location

```python
emp_df2 = pd.read_csv('Emp_data.csv').head()
emp_df2
```

```python
'Date of Birth' = ['27-05-1993','23-03-1995','10-08-1990','25-08-1985','16-08-1998']
'Experience' = [5,4,8,12,6]
```

```python
# Inserting two cols at a time at location - 2

cols = ['Date of Birth', 'Experience']
cols_values = [['27-05-1993','23-03-1995','10-08-1990','25-08-1985','16-08-1998'],[5,4,8,12,6]]
k = 2
for i, col in enumerate(cols):
    for j, col_val in enumerate(cols_values):
        if j == i:
            emp_df2.insert(loc=k, column=col, value=col_val)
            k += 1
```
