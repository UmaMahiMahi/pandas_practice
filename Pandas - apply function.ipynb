```python
data = {"Cust ID": ['C1001','C1005','C1050','C2500'],
        "Principal": [20000,40000,25000,30000],
       'Rate' : [8.9,10.45,7.6,9.5],
       'Time' : [3,4,5,2]}

df2 = pd.DataFrame(data)
df2
```

```python
import datetime
def clean_date(date):
    try:
        new_date = datetime.datetime.strptime(date,'%d-%m-%Y')
        
    except:
        new_date = datetime.datetime.strptime(date,'%d/%m/%Y')
        
    return new_date

df2['Date'].apply(clean_date)
```

```python
import re

string = string.replace("u'",'')
string = re.sub('[^a-zA-Z,]', ' ', string)
string_list = string.split(',')

string_list = list(map(lambda x : x.strip(),string_list ))
string_list[0]
```

```python
movies_df = pd.read_csv('Movies_Data.csv')
def get_actors_list(string):
    string = string.replace("u'",'')
    string = re.sub('[^a-zA-Z,]', ' ', string)
    string_list = string.split(',')

    string_list = list(map(lambda x : x.strip(),string_list ))
    return string_list

movies_df['actors_list'] = movies_df['actors_list'].apply(get_actors_list)
movies_df.dtypes
```
