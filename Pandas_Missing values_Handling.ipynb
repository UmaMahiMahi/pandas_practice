## Dropping Missing Values

```python
Functions to Drop Missing Values:
    1. df.dropna(thresh = col_count, axis = 0, inplace = False)
       df.dropna(thresh = row_count, axis = 1, inplace = False)
    
    2. df.drop('col_name', axis = 1, inplace = False)
    3. df.drop([col_1,col_2,..col_N], axis = 1, inplace = True/False)
    
    4. df.drop('row_name', axis = 0, inplace = False)
       df.drop('row_name', inplace = False)

    5. df.drop([row_1,row_2...row_N], axis =0)
    
```

```python
Syntax:
    df.dropna(axis, how, thresh, subset, inplace)
```

### Parameters in Dropping Null Values:

```python
1. axis:          # Default Value --> 0    ( means row)
    0         --> Removes 'rows' that contain NULL Value
    1         --> Removes columns that contain NULL Value
    'index'   --> Removes 'rows' that contain NULL Value
    'column'  --> Removes columns that contain NULL Value

2. thresh:
    Number   --> No. of NOT NULL values to be required in the 'row'
    
3. inplace:      # Default Value --> False
    False    --> Drops the NULL values in a copied DF (Original DF will not get effect)
    True     --> Drops the NULL values in Original DF and update it

4. index:
    String   --> specifies single index (row) to drop
    List     --> specifies multiple indexes (rows) (Passed in the List) to drop 
5. columns:
    String   --> specifies single column to drop
    List     --> specifies multiple columns (passed in the List) to drop 

6. how:          # Default Value -- "any"
    'all'   --> Specifies whether to drop the row/col when all values are NULL
    'any'   --> Specifies whether to drop the row/col when any values are NULL
7. subset:
    List    --> Specifies where to look for NULL values in columns List

8. errors:       # Default value -- 'ignore'
    'igonre' --> Specifies whether to ignore errors
    'raise'  --> Specifies whether to raise errors
    
    
df.dropna(parameters):
    parameters: thresh, axis, inplace, subset, how
df.drop(parameters):
    parameters: axis, inplace, index, columns, errors
```
