### Advantages of Pandas

```python
1. Fast and efficient for manipulating and analyzing data.
2. Data from different file objects can be loaded (csv, JSON, xlsx, DAT).
3. Easy handling of missing data (represented as NaN) in floating point as well as 
   non-floating point data
4. Size mutability: columns can be inserted and deleted from DataFrame 
    and higher dimensional objects
5. Data set merging and joining.
6. Flexible reshaping and pivoting of data sets
7. Provides time-series functionality.
8. Powerful "group by" functionality for performing split-apply-combine operations 
   on data sets.
```

## Data Structures

```python
Two Types:
    1. Series     --> One Dimensional Data Structure - 1D       >> column data
    2. Data Frame --> Two Dimentional Data Structure - 2D       >> rows & columns as Table
```

#### 1. Series Data Structure

```python
Series: 
    - Pandas Series is One Dimensional labelled array holding the data of any data type. 
      (int, float, str, python object etc)
    - It is Column in Excel.

    Rows   - Indexes starts from 0 by default
    Column - Data

Syntax:
    s1 = pd.Series(data, index = None, dtype = None)
    
data: one dimentional array, list, tuple, dict (key --> row labels, value --> col values)
index: row labels (optional)
dtype: inferred from data (optional)

Default DataType of Series is: float         Futurewarning will occur
        Note: if mention dtype = object      Futurewarning will go off.
        
Datatype of Pandas Series Object:
    pandas.core.series.Series
```

```python
# Creating Series using User Defined Functions

# Fibonacci Series 

def fibo(x):
    if x == 0:
        return x
    elif x == 1:
        return x
    else:
        return fibo(x-1)+fibo(x-2)
n = 10 
print(f'Fibonaci Series of {n} terms:')
for i in range(n):
    x = fibo(i)
    print(x, end = ' ')
s1 = pd.Series([fibo(i) for i in range(n)], index = list(range(1,11)))
s1
```

#### 2. Data Frame

```python
Data Frame: 
    DataFrame is Two Dimensional Data Structure.
    Data is aligned in Tabular Format of rows and columns

syntax:
    df = pd.DataFrame(data, index = None, columns = None, dtype = None)

data: rows and columns --> can be list, tuple, dict (keys--> col labels, values -->records)
index: row labels (optional), default start from 0
columns: col labels (optional), default start from 0
dtype: None (optional) 

Data Frame has 3 Components:
    1. Rows    >> also called as index
    2. Columns
    3. Data    >> value present in the cell

Datatype of Data Frame:
    pandas.core.frame.DataFrame
```

```python
array = np.random.randint(10,50, size = (4,6))
print(array, 'Dimention:',array.ndim)
```

```python
df1 = pd.DataFrame(array)
df1
```

```python
# columns names can be changed

df1 = pd.DataFrame(array, columns = ['A','B','C','D','E','F'])
df1
```

```python
# calling of specific column
# Results in series if No. of Columns = 1

df2 = df1['B']
print(df2)
print(type(df2))
```

```python
# calling of specific column
# Results in DataFrame if No. of Columns > 1

df2 = df1[['A','B']]                 # multiple columns shall be passed in List
print(df2)
print(type(df2))
```

### Examples on Data Frames

#### Creating Fibonacci Series in DF

```python
# Data Frame of Fibo Series of 20 No.s in (2,10) shape
def fibo(x):
    if x == 0:
        return x
    elif x == 1:
        return x
    else:
        return fibo(x-1)+fibo(x-2)
n = 20
fibo_list = [fibo(i) for i in range(n)]              # Fibo list
fibo_array = np.array(fibo_list)                     # Fibo list converted into Np Array
fibo_df = pd.DataFrame(fibo_array.reshape(2,10))     # Data Frame of size (2,10)
fibo_df
```

### Comparision of Series and Data Frame in 1-Dimension

```python
Series:
    1-D  >> one column with its values
Data Frame:
    2-D  >> row and column
    
Single Column DF can be converted into a Series
```

```python
x = np.arange(100,110)
df = pd.DataFrame(x)                      # Data Frame with Single Column
print(df)
print(type(df))                           # frame.DataFrame
```

```python
x = np.arange(100,110)
df = pd.Series(x)                      # Series with Single Column
print(df)
print(type(df))                        # series.Series'
```

```python
var = np.arange(10,25,2)
var_series = pd.Series(var)
var_df = pd.DataFrame(var)
print('Variable Series is:\n', var_series)
print('Variable Data Frame is:\n', var_df)
```

#### Modifications of Columns and its values:

```python
If column already exists: 
    Original records will be replace with New Values
If column not exist:
    New column will be added at the end
```

```python
# To replace the original records
dict1 = {'A': [10,20,30,40],
         'B': (2,3,5,7),
         'C': np.array(['p',4,6,8]),
         'D': np.arange(101,105),
         'E': pd.Series(['p','y','t','h'])
        }
df = pd.DataFrame(dict1)
print('Original DataFrame:')
print(df)
df['C'] = ('x','y','z','k')
print('Updated DataFrame:')
print(df)                        # C-column will be updated
```

```python
# To add one more column to the Data Frame

df['F'] = pd.Series([1,8,27,64])
df                                 # F' column will be added
```

## Reading Data from Files:           Input/Output

```python
1. CSV:
    To Read: 
        pd.read_csv(r'filepath\file_name.csv')
    To Convert DF:
        df.to_csv(r'filepath\file_name.csv')
        
2. Excel:
    To Read:
        pd.read_excel(r'filepath\file_name.xlsx')
    To Convert DF:
        df.to_excel(r'filepath\file_name.xlsx')
        
3. Pickle:
    To Read:
        pd.read_pkl(r'filepath\file_name.pkl')
    To Convert DF:
        df.to_pickle(r'filepath\file_name.pkl')
        
4. JSON:
    To Read:
        pd.read_json(r'filepath\file_name.json')
    To Convert DF:
        df.to_html(r'filepath\file_name.json')
        
5. HTML:
    To Read:
        pd.read_html(r'filepath\file_name.html')
    To Convert DF:
        df.to_html(r'filepath\file_name.html')
        
6. SQL:
    To Read:
        pd.read_sql_table(r'filepath\file_name.sql', con)       # read SQL Table
        pd.read_sql_query(r'filepath\file_name.sql', con)       # read SQL Query Database
        pd.read_sql(r'filepath\file_name.sql', con)             # read SQL Table or Query DB
    To Convert DF:
        df.to_sql(r'filepath\file_name.sql')
        
    Note: con >> Database Configuration should specify to read SQL file
7. STATA:
    To Read:
        pd.read_stata(r'filepath\file_name.dta')
    To Convert DF:
        df.to_stata(r'filepath\file_name.dta')
```
