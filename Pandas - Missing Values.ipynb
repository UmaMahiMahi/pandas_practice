### Pandas - Missing Values

```python
import pandas as pd
import numpy as np
```

### Functions

```python
1. Detect Missing Values:
    1. df.isna()
    2. df.isnull()
    3. df.isna().sum()
    4. df.isna().mean()
    5. df.isna().mean() * 100
    
2. Drop Null Values:
    1. df.dropna():
        1. df.dropna(thresh=df.shape[1], axis = 0)
        2. df.dropna(thresh=df.shape[0], axis = 1)
        
    2. df.drop():
        df.drop('ColName', axis = 1)
        df.drop(['Col1', 'Col2', ...,'ColN'], axis = 1)
        df.drop(['Row1', 'Row2',..., 'RowN'], axis = 0)
    
    
3. Imputation:
    1. df.fillna()
        Replace Missing values by using Stats:

        Continuous Data:
            1. df['ColName'].mean()
            2. df['ColName'].median()

        Categorical Data:
            1. df['ColName'].mode()[0]
```
