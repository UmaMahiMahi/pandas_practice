```python
movies_df = pd.read_csv('movies_data.csv')
movies_df
```

```python
action_df = movies_df.groupby('genre').get_group('Action')
action_df
crime_df = movies_df.groupby('genre').get_group('Crime')
crime_df
```

```python
movies_df.to_excel("Movies.xlsx", sheet_name='All_Movies',index = False)
```

```python
action_df.to_excel("Movies.xlsx", sheet_name='Action',index = False)
```

```python
crime_df.to_excel("Movies.xlsx", sheet_name='Crime',index = False)
```

```python
Writer = pd.ExcelWriter('Movies.xlsx', engine='xlsxwriter')     #Class for writing DataFrame objects into excel sheets.


movies_df.to_excel(Writer, sheet_name='All_Movies',index = False)
action_df.to_excel(Writer, sheet_name='Action',index = False)
crime_df.to_excel(Writer, sheet_name='Crime',index = False)
Writer.save()
```
