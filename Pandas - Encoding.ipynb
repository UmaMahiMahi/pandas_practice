### Encoding

```python
Data:
    Categorical Data 
    Continuous Data
    
Categorical Data:
    1. Ordinal Data >> Label Encoding
    2. Nominal Data >> One Hot Encoding
    
Ordinal Data(Data will be in order)
    CreditRisk:
        High
        Medium
        Low
    
    Education:
        GRADUATE
        UNDER GRADUATE
        POST-GRADUATE
        HSC
        SSC
        PHD
    
Nominal Data:(Data will not be in order)
    southwest', 'southeast', 'northwest', 'northeast'
    City Names: Pune, Hyderabad, Delhi, Mumbai
        
        
```
