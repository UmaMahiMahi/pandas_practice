### pandas - where

```python
import pandas as pd
import numpy as np
```

```python
np.where(condition)
```

```python
array = np.random.randint(10,30,size = 10)
array
```

```python
index = np.where(array >= 20)[0]
index
```

```python
array[index]
```

```python
array = np.random.randint(10,30,size = 10)
numbers = array[np.where(array >= 20)[0]]
numbers
```

```python
numbers = np.where(array >= 20,200,100)
numbers
```

```python
numbers = np.where(array >= 20,1,0)
numbers
```

```python
numbers = np.where(array >= 20,'T','F')
numbers
```

```python
numbers = np.where(array >= 20,200,array)
numbers
```

```python
numbers = np.where(array >= 20,array,0)
numbers
```

```python
emp_df = pd.read_csv('Emp_data.csv').head(10)
emp_df
```

```python
np.where(emp_df['Salary'] >= 100000,'High', 'Low')
```

```python
emp_df['Salary'] = np.where(emp_df['Salary'] >= 100000,'High', 'Low')
emp_df
```

```python
emp_df = pd.read_csv('Emp_data.csv').head(10)
emp_df
```

```python
emp_df = pd.read_csv('Emp_data.csv').head(10)
emp_df['Salary'] = np.where(emp_df['Salary'] >= 150000, 'High', 
                            np.where(emp_df['Salary'] >= 90000,'Medium','Low'))

emp_df
```
