## Pandas Functions

```python
1. df.columns     >> List of columns 
2. df.index       >> Index range: start, stop, step
3. df.axes        >> List of index, columns
4. df.dtypes      >> Series of columns datatypes
5. df.select_dtypes(include, exclude)    >> Columns of particular Datatypes
6. df.info()      >> info of the dataset: col labels, null, non-null values, dtypes
7. df.describe()  >> Descriptive Stats: count, std, median, quantiles, min, max
8. df.shape()     >> shape of the DataFrame


Accessing of Data:

Accessing:
    df['col_name']             >> Series
    df[['col_name']]           >> DataFrame
    df[['col1, col2...colN']]  >> DataFrame

Index Based Location: iloc --> index location
    Series:
        s1.iloc[index_value]
    DataFrame:
        df.iloc[row_index, col_index]
        df.iloc[row_index_start : row_index_end, col_index_start : col_index_end]]
        df.iloc[row_index_list, col_index_list]
    Note: end parameter excludes
        
Label Based Location: loc --> label location
    Series:
        s1.loc[row_label]
    DataFrame:
        df.loc[row_label, col_label]
        df.loc[row_label_start : row_label_end, col_label_start : col_label_end]
        df.loc[row_label_list, col_label_list]
    Note: end parameter includes
        
        
Row Indexes:
    df[row_index_start : row_index_end]   >> able to access the rows only
    
Column Labels:
    df.col_name
    Note: No spaces presents inbetween col name (underscore is allowed)
        
N - Rows:
    1. df.head(N)   >> First N- rows  (Default Value: 5)
    2. df.tail(N)   >> Last N- rows   (Default Value: 5)
```

### 2. Index-Location     --- iloc

```python
To access index-locations:
    df.iloc[row_index, col_index]              
    df.iloc[row_index]                         
    df.iloc[row_index_start:row_index_end , col_index_start:col_index_end]
    df.iloc[row_index_list, col_index_list]   
    
    
Notes:
    df.iloc[:,:]      --> all values
    df.iloc[x:y,a:b]  --> rows (x to y) and cols (a to b)
    df.iloc[x: , a]   --> rows (x to end) and col (a)
    df.iloc[x: , a: ] --> rows (x to end) and cols (a to end)
    df.iloc[x: , :b]  --> rows (x to end) and cols (0 to b)
    df.iloc[x: , : ]  --> rows (x to end) and all cols
    df.iloc[:y , a]   --> rows (0 to y) and col (a)
    df.iloc[:y , a: ] --> rows (0 to y) and cols (a to end)
    df.iloc[:y , :b]  --> rows (0 to y) and cols (0 to b)
    df.iloc[:y , : ]  --> rows (0 to y) and all cols
    df.iloc[: , a: ]  --> all rows and cols (a to end)
    df.iloc[: , :b]   --> all rows and cols (0 to b)
    
    df.iloc[[x,y,z], [a,b,c]] --> rows(x,y,z) and cols (a,b,c)
    
    Here: x,y,z and a,b,c are index values
          End Parameter will include in the result
```

### 3. Location_Label               --- loc

```python
Access with Label_Names:
    1. df.loc[row_labels, col_labels]
    2. df.loc[row_labels]
    3. df.loc[:, col_labels]
    
    Note: start and end labels will be included
    
Notes:
    df.loc[:,:]      --> all rows & cols
    df.loc[x:y,a:b]  --> rows (x to y) and cols (a to b)
    df.loc[x: , a]   --> rows (x to end) and col (a)
    df.loc[x: , a: ] --> rows (x to end) and cols (a to end)
    df.loc[x: , :b]  --> rows (x to end) and cols (0 to b)
    df.loc[x: , : ]  --> rows (x to end) and all cols
    df.loc[:y , a]   --> rows (0 to y) and col (a)
    df.loc[:y , a: ] --> rows (0 to y) and cols (a to end)
    df.loc[:y , :b]  --> rows (0 to y) and cols (0 to b)
    df.loc[:y , : ]  --> rows (0 to y) and all cols
    df.loc[: , a: ]  --> all rows and cols (a to end)
    df.loc[: , :b]   --> all rows and cols (0 to b)
    
    df.loc[[x,y,z], [a,b,c]] --> rows(x,y,z) and cols (a,b,c)
    
    Here: x,y,z are row labels and a,b,c are column labels
    End Parameter will include in the result
```
